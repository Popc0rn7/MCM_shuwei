```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
hainan = pd.read_csv('hainan_output.csv')
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
print(hainan)
print(iris)
print(tips)
```

## catplot(category)

```python
sns.set_theme(style="whitegrid",font_scale=1,palette="Set2")
sns.displot(data=iris,x='sepal_length',hue='species',multiple='stack',col='species')
```

```python
sns.pairplot(iris,hue='species')
sns.jointplot(data=iris,x='sepal_width',y='sepal_length',hue='species',kind='hist')
```

```python
#sns.lineplot(data=iris,x='sepal_width',y='sepal_length',hue='species')
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',row='smoker',col='time')
```

```python
sns.regplot(data=iris,x='sepal_width',y='sepal_length')
flights_long = sns.load_dataset("flights")
print(flights_long)
```
